```python
# 6. View Recorded Game Videos
import os, glob
from IPython.display import Video, display

video_dir = '/content/drive/MyDrive/quoridor_videos/'
video_files = sorted(glob.glob(f'{video_dir}/*.webm'), key=os.path.getctime)

if video_files:
    print(f'Found {len(video_files)} video(s). Showing the latest:')
    display(Video(video_files[-1], embed=True, width=800))
    print('\nAll recorded videos:')
    for v in video_files:
        print(' -', os.path.basename(v))
else:
    print('No videos found. Check record_video_dir in settings.yaml.')
```
